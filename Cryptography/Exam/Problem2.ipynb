```python
p = 2
n = 4
val = 1
for i in range(n):
    val = val*p
    
for i in range(val):
    print(bin(i)[2:])
```

```python
table = []

for i in range(val):
    for j in range(val):
        table.append(bin((i+j) % val)[2:])
        
for i in range(len(table)):
    print(table[i]," ",end='')
    if((i+1)%val == 0):
        print('\n')

add1 = 16
add2 = 12

print(bin((add1+add2)%val)[2:])
```
